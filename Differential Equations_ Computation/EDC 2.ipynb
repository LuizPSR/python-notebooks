# Questão 1
Um sistema mola-massa tem seu pode ser descrito como a seginte equação em derivada do tempo:

m x'' + kx = Fext

m = 1kg, k = 4N/m

a) Sendo Fext = 0, plote o grafico da solução de x ao longo do tempo

-> Solução

  x' = y

  y' = 4x 

  [x]' = [ 0  1] [x]

  [y]  = [-4  0] [y]

  lambda = 2i

  x = c1 sen 2t + c2 cos 2t

```python
t = sym.symbols('t')
xh = sin(2*t) + cos(2*t)
sym.plot(xh, (t,0,10), ylabel = 'x')
```

b) Plote o grafico para Fext = 10cos(2.1t)

-> Solução

  Chute:
  
  x  = Acos(2.1t) + Bsin(2.1t)

  x' = -2.1Asin + 2.1Bcos

  x" =  -4.41A cos - 4.41B sin

  4x + x" = 10 cos

  4*A - 4.41A = 10
  
  A = 10/0.41

```python
xp = -10/0.41*cos(2.1*t)
sym.plot(xh+xp, (t,0,100), ylabel = 'x')
```

c) E se Fext = 10cos(2t)?

-> Solução

  Chute:

  x  = At cos 2t + Bt sin 2t

  x' = Acos - 2Atsin + Bsin + 2Btcos

  x" = -2Asin - 2Asin - 4Atcos + 2Bcos + 2Bcos - 4Btsin

  x" = sin (-4A - 4Bt) + cos ( -4At + 4B)

  4x + x" = 10 cos

  A = 0, B = 10 / 4

```python
xp = 2.5*t*sin(2*t)
sym.plot(xh+xp, (t,0,200), ylabel = 'x')
```

# Questão 2
Um circuito RLC pode ser escrita como:

Li" + Ri' + i/C = E'

em que i ́e a corrente no sistema, L ́e a indutância,R ́e a resistência, C ́e a capacitância e E ́e a fonte de tensão.

Considerando L = 1H, R = 0.25Ω e C = 4F.

a) Plote i ao longo do tempo se i(0) = 1 e i'(0) = 0. (Na falta de informação, considerei uma Tensão constante, ou seja E' = 0)

-> Solução

  em t0. E' = 1/4 + i"

  [i'] = [  0    1 ][i]

  [j'] = [-1/4 -1/4][j]

  1/4 + l/4 + l² = 0

  l = -1/8 +/- i sqrt(15)/8 

  i = e^(-t/8) (c1 cos t * sqrt(15)/8 + c2 sen t * sqrt(15)/8)

  i(0) = 1 -> c1 = 1

  i'(0) = 0 = -1/8 * e^(-t/8) * (c1 cos + c2 sin + sqrt(15)*(c1 sin - c2 cos))

  i'(0) -> -1/8 * (c1 - sqrt(15) c2)

  c2 = sqrt(15)/15

```python
t = sym.symbols('t')
ih = np.e ** (-t/8) * (1 * cos(t*np.sqrt(15)/8) + (np.sqrt(15)/15) * sin(t*np.sqrt(15)/8))
sym.plot(ih, (t,0,50), ylabel = 'i')
```

b) Considere agora uma fonte externa na forma 

E (t) = 10cos(ωt),

 o que implica na relação

 E' (t) = −10ωsen (ωt)
 
 _

Plote a corrente ao longo do tempo, considerando condições iniciais nulas e para valores de ω ∈ {0.05, 0.2, 4}. Qual deles você acredita estar mais
próximo da frequência de ressonância? Por quê?

-> Solução

  usando a homologa encontrada na letra a, acharemos agora uma solução particular

  Chute:  
  
  i  = A sin(ωt) + B cos(ωt)

  i' = Aω cos(ωt) - Bω sin(ωt)

  i" = -Aω² sin(ωt) - Bω² cos(ωt)

  i" + i'/4 + i/4 = -10ω sin(ωt)

  Aω + B - Bω² = 0

  A - Bω - Aω² = -10ω

___

  ω = 0.05 -> A = -9.9/4.913, B = 1/9.826
  
  ω = 0.2  -> A = -2100/233,  B = 500/233

  ω = 4    -> A = 2016/797,   B = 128/797

```python
ω = 0.05
A = -9.9/4.913 # ~ -2.01
B = 1/9.826    # ~  0.10
ip = A * sin(ω*t) + B * cos(ω*t)
sym.plot(ih+ip, (t,0,200), ylabel='i')
```

```python
ω = 0.2
A = -2100/233 # ~ -9.01
B = 500/233   # ~  2.14
ip = A * sin(ω*t) + B * cos(ω*t)
sym.plot(ih+ip, (t,0,100), ylabel='i')
```

```python
ω = 4
A = 2016/797 # ~ 2.53
B = 128/797  # ~ 0.16
ip = A * sin(ω*t) + B * cos(ω*t)
sym.plot(ih+ip, (t,0,50), ylabel='i')
```

A situção mais próxima de ressonância me parece ocorrer quando ω = 4.
